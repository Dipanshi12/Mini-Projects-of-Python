```python
menu = {
    "Burger" : 45,
    "Pizza" : 100,
    "Cold drink" : 50,
    "Coffee" : 35,
    "Noodles" : 150
}

print("-----MENU-----")
for item,price in menu.items():
    print(f"{item} : Rs = {price}")
name =input("\n enter name:")
mobile = int(input("\n enter mobile number:"))
address = input("\n enter address:")
def generate_bill(*args, **kwargs):
    print("\n-----BILL-----")
    print(f"Customer_name : {kwargs['Name']}")
    print(f"Customer_mobile : {kwargs['Mobile']}")
    print(f"Customer_address : {kwargs['Address']}")

    total = 0
    print("\n items ordered:")
    for item in args:
        print(f"{item} : Rs. {menu[item]}")
        total +=menu[item]
        
    gst = total *0.18
    final_amount = total + gst

    print("\n---------------")
    print(f"subtotal : {total}")
    print(f"Gst(18%) : Rs.{gst:.2f}")
    print(f"final_amount : Rs.{final_amount}")

orders = []
while True:
    item = input("\n enter item name (type 'done' to finish):")
    if item.lower() == "done":
        break
    if item in menu:
        orders.append(item)
    else:
        print("Item not available!")
generate_bill(
    *orders,
    Name = name,
    Mobile = mobile,
    Address = address
)
    
```
