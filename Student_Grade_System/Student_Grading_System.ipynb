```python
#Dictionary
Students = {}
#Function adding student
def Add_Student():
    Name = input("enter name of student:")
    Marks = int(input("enter marks:"))
    Students[Name] = Marks
#Function displaying 
def Display_Student():
 for Name,Marks in Students.items():
    print(f"{Name} : {Marks}")
#Function searching
def Search_Student():
    Name = input("Student search:")
    if Name in Students:
        print("Marks:", Students[Name])
    else:
        print("Student not found")
#Function calculating grade
def Calculate_Grade():
    Name = input("Student grade:")
    if Name in Students:
     Marks = Students[Name]
     if Marks >= 90:
        print("Grade A")
     elif Marks >= 80:
        print("Grade B")
     elif Marks >= 70:
        print("Grade C")
     elif Marks >= 60:
        print("Grade D")
     else:
        print("Fail")
    else:
     print("Student not found")
#Function menu
def Menu():
    while True:
        print("\n-----Grading-----")
        print("1.Add student")
        print("2.Display student")
        print("3.Search student")
        print("4.Calculate grade")
        print("5.Exit")
        choice = int(input("Enter choice"))
        if choice == 1:
            Add_Student()
        elif choice == 2:
            Display_Student()
        elif choice == 3:
            Search_Student()
        elif choice == 4:
            Calculate_Grade()
        elif choice == 5:
            print("Exit")
            break
        else:
            print("invalid choice")
#Calling function            
Menu()
```
