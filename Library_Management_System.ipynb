```python
#Library_Management_System
books = {
    "python":5,
    "java":8,
    "c":1,
    "maths":3
        }
#Display
def display():
    print("\n Available books")
    for book,quantity in books.items():
        print(f"{book} : {quantity} copies")
#Add_book
def add():
  book_name = input("enter book to add:")
  if book_name in books:
   books[book_name] +=1
   print("book quantity updated")
  else:
   books[book_name] = 1
   print("book added")
#Issue_book
def issue():
  book_name = input("enter book to issue:")   
  if book_name in books:
    if books[book_name] > 0:
       books[book_name] -=1
       print("Book issued ")
    else:
       print("sorry! book not available")
  else:
      print("book not found")
#Return_book
def return_book():
    book_name = input("enter book to return:")
    if book_name in books:
        books[book_name] += 1
    else:
        books[book_name] = 1
        print("book returned successfully")
#Main_menu
def library():
 while True:
  print("\n-----Library books menu-----")
  print("1.display book")
  print("2.add book")
  print("3.issue book")
  print("4.return book")
  print("5.exit")
  choice = int(input("enter choice:"))
  if choice == 1:
     display()
  if choice == 2:
     add()
  if choice == 3:
     issue()
  if choice == 4:
     return_book()
  if choice == 5:
     print("Thanks!")
     break
  else:
    print("invalid choice")

library()
```
